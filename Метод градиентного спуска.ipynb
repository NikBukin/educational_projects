```python
def F(x):
    return 2*x[0]**2 + 5*x[1]**2 + x[0]*x[1] - 3*x[0] - x[1]
def grad(x):
    return [4*x[0] + x[1] - 3 , 10*x[1] + x[0] - 1]
```

```python
def mingrad(a,e):
    x = [1,1]
    z = [0,0]
    z[0] = x[0] - a*grad(x)[0]
    z[1] = x[1] - a*grad(x)[1]
    i = 0
    while (grad(x)[0]**2 + grad(x)[1]**2)**(1/2)>e:
        z[0] = x[0] - a*grad(x)[0]
        z[1] = x[1] - a*grad(x)[1]
        if F(z)>F(x): a = a/2
        if F(z)<F(x): x = z
        i = i + 1
        print([i, a, x, F(x),(grad(x)[0]**2 + grad(x)[1]**2)**(1/2)])

    print('')
    print('Полученное оптимальное значение х* = ', x)
    print('Соответствующее значение f(х*) = ', F(x))
```

```python
mingrad(1,0.0001)
```
